```python
def removeSentenceWithoutVerb(fileContent):
    ''' Removes sentences from the file which do not have any verb forms '''
    #print (fileContent)
    fileData = []
    for eachSentence in fileContent:
        sent = nlp(eachSentence)
        isVBPresent = False
        for j in sent:
            if j.tag_.startswith("VB"): isVBPresent=True
        if isVBPresent:
            fileData.append(eachSentence)
    #input()
    return fileData
```

```python
def dataloader(path):
    ''' Parses the data 
        1. Assumed sentences end with '. '
        2. Cleans sentences - strips spaces
        3. Remove sentences if no verb form
    '''
    
    allFileContents = {}
    files = [ f for f in listdir(path) if f.endswith(".txt")]
    #files = ['2.txt']
#     print(files)
    for f in tqdm(files):
        try :
            base = os.path.basename(f).split('.')[0]
            ff = os.path.join(path,f)
            fileHandle = open(ff,'r',encoding="utf-8")
    #         fileHandle = open(ff,'rb')
            fw =  open(os.path.join('web_annotations/',base)+'.csv','w')
            writer = csv.writer(fw, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)

            data = fileHandle.read().split('\n')
            text = [i.replace("\t"," ").replace('“',"'").replace('”',"'").replace("…","...").replace("’","'").strip("\n").strip(" ").split(". ") for i in data]
            #print (text)
            fileContent = []
            for each in text:
                for eacheach in each:
                    if eacheach and len(eacheach)>2:
                        fileContent.append(eacheach.strip(" "))
            '''if base == '13':
                print (fileContent)
                input()'''

            allFileContents[base] = {"content":removeSentenceWithoutVerb(fileContent)}
            #print (allFileContents )
            #input()

            with open('ctfwriteup_bin.json','w') as fw:
                json.dump(allFileContents, fw)
            '''
            with open('ctfwriteup.jsonl','w') as fw:
                fw.write(allFileContents)'''
            for lineno,eachline in enumerate(allFileContents[base]['content'],1):
                writer.writerow([base,lineno,eachline])
        
        except:
            print (f)
    return allFileContents
        
```

```python

actionRules = ["PRP-MD-VB","PRP-RB-VB","PRP-VB","PRP-MD-RB-VB"]
infoRules = ["NN-VB-NN","NN-VB-VB","NN-VB-RB-NN","NN-RB-VB-NN","NN-MD-VB","NN-MD-RB-VB","EX-VB"]
filterPos = 'PDT|DT|IN|,|:|\.|JJ|SYM|TO|HYPH'
```

```python
def getTags(datadict):
    
    moddict = {}
    for eachfile in tqdm(datadict):
        all_mod_sent = []
        for linenum, eachline in enumerate(datadict[eachfile]['content']):

#             print (linenum, eachline)
#             input()
            posTag = []
            for word in nlp(eachline):
                posTag.append(word.tag_)
            posTagSeq = "-".join(posTag)

            reducedPosTagSeq = [x for x in posTag if not re.search(filterPos,x) ]
            reducedPosTagSeqMod = []
            for x in reducedPosTagSeq:
                if re.match("NN*",x):
                    reducedPosTagSeqMod.append("NN")
                elif re.match("VB*",x):
                    reducedPosTagSeqMod.append("VB")
                else:
                    reducedPosTagSeqMod.append(x)

            posTagSeqNew = "-".join(reducedPosTagSeqMod)

            predEnttypeAction = ''
            for eachRule in actionRules:
                if eachRule in posTagSeqNew:
                    predEnttypeAction = 'A'

            predEnttypeInfo = ''
            for eachRule in infoRules:
                if eachRule in posTagSeqNew:
                    predEnttypeInfo = 'I'
    #                 print ("Rule:",eachRule)
            sent_type = ""
            if predEnttypeAction and predEnttypeInfo:
                sent_type = 'A/I'
            elif predEnttypeAction:
                sent_type = "A"
            elif predEnttypeInfo:
                sent_type = "I"
            else:
                sent_type = 'O'

            all_mod_sent.append((linenum, eachline, sent_type))
        moddict[eachfile]=all_mod_sent
    return moddict
```
